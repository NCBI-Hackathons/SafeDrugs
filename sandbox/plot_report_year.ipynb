```python
def plot_report_year():
    """
    """
    
    home_path = os.path.abspath("..")
    data_path = os.path.join(home_path, "data", "aeolus_top5drugs.feather")
    df = pd.read_feather(data_path)
    
    mpl.style.use('seaborn-whitegrid')
    mpl.rcParams['font.weight']= 'bold'
    mpl.rcParams['font.size']= 16
    
    data_group_by_y = df.groupby("report_year")["id"].count()
    
    _, ax = plt.subplots(figsize=(10, 10))
    
    p = data_group_by_y.plot(kind = "bar", ax=ax)
    p = ax.set_xlabel("Report Year")
    p = ax.set_ylabel("Reports")
    p = ax.set_xticklabels(ax.get_xticklabels(),rotation=45)
    

    
    
    
```

```python
plot_report_year()
```
