```python
import pandas as pd
%matplotlib inline
import matplotlib as mpl
from quilt.data.hsiaoyi0504 import aeolus_top5drugs


mpl.style.use('seaborn-whitegrid')
mpl.rcParams['font.weight']= 'bold'
mpl.rcParams['font.size']= 16
             
data = pd.read_feather(aeolus_top5drugs.aeolus_top5drugs._data())
```

```python
grpd = data.groupby('drug_concept_id')['id'].count()
```

```python
_,ax = mpl.pyplot.subplots(figsize=(10,10))
p = grpd.plot(kind='bar',ax=ax)
p = ax.set_ylabel('Number of reports')
p = ax.set_xlabel('Drug')
p = ax.set_xticklabels(ax.get_xticklabels(),rotation=45)
```
