```python
import matplotlib as mpl
%matplotlib inline
import matplotlib.pyplot as plt

import pandas as pd
import numpy as np
import os
```

```python
def plot_report_year():
    """
    plot number of reports with respect to the years.
    """
    
    home_path = os.path.abspath("..")
    data_path = os.path.join(home_path, "data", "aeolus_top5drugs.feather")
    df = pd.read_feather(data_path)
    
    mpl.style.use('seaborn-whitegrid')
    mpl.rcParams['font.weight']= 'bold'
    mpl.rcParams['font.size']= 16
    
    data_group_by_y = df.groupby("report_year")["id"].count()
    
    _, ax = plt.subplots(figsize=(10, 10))
    
    p = data_group_by_y.plot(kind = "bar", ax=ax)
    p = ax.set_xlabel("Report Year")
    p = ax.set_ylabel("Reports")
    p = ax.set_xticklabels(ax.get_xticklabels(),rotation=45)
    

    
    
    
```

```python
plot_report_year()
```

```python
def n_drugs_report(years = ['2014']):
    """
    plot number of each drug reported with respect to the drug names.
    """
    
    home_path = os.path.abspath("..")
    data_path = os.path.join(home_path, "data", "aeolus_top5drugs.feather")
    df = pd.read_feather(data_path)
    
    mpl.style.use('seaborn-whitegrid')
    mpl.rcParams['font.weight']= 'bold'
    mpl.rcParams['font.size']= 16
    
    data_of_years = []
    for year in years:
        data_of_years.append(df.query('report_year==@year'))
    
    all_data = pd.concat(data_of_years)
    all_data = all_data.groupby("drug_concept_name")["id"].count()
    
    _, ax = plt.subplots(figsize=(10, 10))
    
    p = all_data.plot(kind = "bar", ax=ax)
    
    years_label = ""
    for year in years:
        years_label += year
    p = ax.set_xlabel("Report Year: {}".format(years_label))
    
    p = ax.set_ylabel("Number of Reports")
    p = ax.set_xticklabels(ax.get_xticklabels(),rotation=45)
```

```python
n_drugs_report()
```
